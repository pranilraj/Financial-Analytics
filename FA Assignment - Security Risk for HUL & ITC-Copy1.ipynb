```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
#The next step is to input trickersymbol of the respective company inorder to fetch the historical datas of the company by
#'yfinance'library. A ticker symbol or stock symbol is an abbreviation used to uniquely identify publicly traded shares of a
#particular stock on a particular stock market. A stock symbol may consist of letters, numbers or a combination of both.
#Then in the next line HUL name has been assigned to fetch history data of the company stock for period of one day.
#coming to the print section yfinance will fetch the recent company stock traded opening price and closing price and all
#necessary details.(similar to ITC)
```

```python
tickerSymbol = 'HINDUNILVR.NS'
tickerData = yf.Ticker(tickerSymbol)
HUL = tickerData.history(period='1d', start='2017-08-01')
print("HUL shares " ,"\n",HUL.tail())

tickerSymbol = 'ITC.NS'
tickerData = yf.Ticker(tickerSymbol)
ITC = tickerData.history(period='1d', start='2017-08-01')
print("\n","ITC shares ","\n" ,ITC.tail())
```

```python
#After having a anlysis on the share price of both the companies the next step is to compare both the company stock prices
#using plot method for that first we have to analyse which byte of data would be most preferable for comparing, in this data
#close price of the shares are most likely to compare and allocating share names for plotting,
#After that we have to create a dataframe of data consisting of the allocated share names and then plotting accordingly setting 
#the figsize 10 as width and 10 as height,
#plt.legend -A chart legend appears by default when you first create a chart. For most charts, legends show the names and colors
#of each series of data. 
#In addition to that setting graph title as mentioned below and setting x- axis or xlabel as 'Year' for comparing the stock
#price(assigned as y-axis or ylabel) of both the companies according to the number of years stock traded.
#And setting all required color and linewidth for plotting.
```

```python
HUL = HUL.rename(columns={'Close' : 'HUL_Shares'})
ITC  = ITC.rename(columns={'Close' : 'ITC_Shares'})
hul_itc_data = pd.DataFrame(data = [HUL['HUL_Shares'],ITC['ITC_Shares']]).T
hul_itc_data.plot(figsize=(10,10))
plt.legend()
plt.title("Adjusted Close Price of Hindustan Unilever Limited & Indian Tobacco Company",fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```python
#Assigning HUL returns a logarithm function inorder to calculate the HUL shares return and shift is used to allign a sapce in 
#between date and the share returns 
#np.log -This mathematical function helps user to calculate Natural logarithm of x where x belongs to all the input
#array elements.
#Finally showing the last 20 share returns of HUL company stocks.
```

```python
HUL_returns = np.log(HUL['HUL_Shares'] / HUL['HUL_Shares'].shift(1))
HUL_returns.tail(20)
```

```python
#Assigning ITC returns a logarithm function inorder to calculate the ITC shares return and shift is used to allign a sapce in 
#between date and the share returns.
#Finally showing the last 20 share returns of ITC company stocks.
```

```python
ITC_returns = np.log(ITC['ITC_Shares'] / ITC['ITC_Shares'].shift(1))
ITC_returns.tail(20)
```

```python
#Now Calculating the mean of HUL company return by using "mean function" which returns the average of HUL share returns.
#And multiplying mean returns by 250
#Second thing is to calculate the standard deviation using the "std function" and then multiplying the standard deviation
#result by 250 and it by 0.5
```

```python
"HINDUSTAN UNIVER LIMITED"

print ("HUL returns mean = ", HUL_returns.mean(),"\n\n",
       "HUL returns mean * 250 = ",HUL_returns.mean()*250,"\n\n",
       "HUL returns Standard Deviation = ",HUL_returns.std(),"\n\n",
       "HUL returns Standard Deviation * 250 * 0.5 = ",HUL_returns.std()*250**0.5)
```

```python
#Now Calculating the mean of ITC company return by using "mean function" which returns the average of ITC share returns.
#And multiplying mean returns by 250
#Second thing is to calculate the standard deviation using the "std function" and then multiplying the standard deviation
#result by 250 and it by 0.5
```

```python
"INDIAN TOBACCO COMPANY"

print ("ITC returns mean = ", ITC_returns.mean(),"\n\n",
       "ITC returns mean * 250 = ",ITC_returns.mean()*250,"\n\n",
       "ITC returns Standard Deviation = ",ITC_returns.std(),"\n\n",
       "ITC returns Standard Deviation * 250 * 0.5 = ",ITC_returns.std()*250**0.5)
```

```python
#Final results of HUL & ITC returns mean value multiplied by 250.
```

```python
"Final Results"

print(HUL_returns.mean()*250)
print(ITC_returns.mean()*250)
```

```python
#Comparing the HUL & ITc company share return mean value and standard deviation value.
```

```python
HUL_ITC_returns = pd.DataFrame(data = [HUL_returns,ITC_returns]).T
print ("HUL vs ITC mean * 250","\n\n",HUL_ITC_returns[['HUL_Shares','ITC_Shares']].mean()*250,"\n\n\n\n",
"HUL vs ITC mean * 250 square 0.5 ","\n\n",HUL_ITC_returns[['HUL_Shares','ITC_Shares']].mean()*250**0.5)
```

```python
#Now final step to plot the returns of  two companies using legend method to plot the chart.
```

```python
HUL_ITC_returns.plot(figsize=(10, 7))
plt.legend()
plt.title("Returns of HUL & ITC ", fontsize=16)
plt.ylabel('Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
